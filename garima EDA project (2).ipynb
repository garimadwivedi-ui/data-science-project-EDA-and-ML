# Iris Flower EDA + Model Project

This notebook performs Exploratory Data Analysis (EDA) and builds a simple ML model on the Iris dataset.

## 📌 AIM

The aim of this project is to analyze the Iris dataset using EDA techniques and build a machine learning model to predict the species of iris flowers based on their features.

## 🎯 OBJECTIVE

- To perform Exploratory Data Analysis (EDA) on the Iris dataset.  
- To understand the relationship between different features of the dataset.  
- To visualize the data using graphs like pairplot, histogram, boxplot, and heatmap.  
- To build a machine learning model for classifying iris flower species.  
- To evaluate the performance of the model using accuracy.  

## 📚 IMPORTING REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 📂 LOADING THE DATASET

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa',1:'versicolor',2:'virginica'})

df.head()
```

## 🔍 BASIC DATA EXPLORATION

```python
df.info()
df.describe()
df.isnull().sum()
```

## 📊 EXPLORATORY DATA ANALYSIS (EDA)

### 📊 Histogram Analysis

```python
df.hist(figsize=(10,8))
plt.show()
```

### 📦 Boxplot Analysis

```python
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.show()
```

### 🔥 Correlation Heatmap

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.show()
```

### 🥧 Species Distribution (Pie Chart)

```python
species_counts = df['species'].value_counts()

plt.figure()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Iris Species")
plt.show()
```

## 🤖 MACHINE LEARNING MODEL
### Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = df.drop('species', axis=1)
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

### Accuracy: 0.9666666666666667

## 🖼️ DISPLAYING FLOWER IMAGE BASED ON PREDICTION

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_flower(species):
    
    if species == 'setosa':
        img = mpimg.imread("setosa.jpg")
    elif species == 'versicolor':
        img = mpimg.imread("versicolor.jpg")
    else:
        img = mpimg.imread("virginica.jpg")
    
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

## 🔮 PREDICTION ON NEW DATA

```python
import pandas as pd

sample = pd.DataFrame(
    [[5.1, 3.5, 1.4, 0.2]],
    columns=X.columns
)

prediction = model.predict(sample)[0]

print("Predicted Flower:", prediction)
show_flower(prediction)
```
